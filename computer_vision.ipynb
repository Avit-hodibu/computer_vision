**threshold function**
A threshold function applies a threshold to every pixel, it can be used in extracting objects from an image this is called segmentation

The following function applies a threshold to the input array input image array imput_out and outputs the result the following portion of the code will cycle through each pixel (,). If the pixel is greater then that threshold. It will set a pixel in the array image_out pixel to some value, usually 1 or 255, Otherwise it will set it to another value, usually 0.
```
def thresholding(input_img, threshold, max_value=255, min_vlaue=0):
  N, M = input_img.shape
  image_out=np.zeros((N,M), dtype =np.uint8)

  for i in range(N):
    for j in range(M):
      if input_img[i,j]> threshold:
        image_out[i,j]= max_value
      else:
        image_out[i,j]= min_value
  return.image_out
```
